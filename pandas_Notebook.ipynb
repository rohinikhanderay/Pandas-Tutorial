# **Pandas**
**Pandas is a Python library.Pandas is used to analyze data.**
Its a fast and efficient dataframe object for data manipulation.
Installation of Pandas
If you have Python and PIP already installed on a system, then installation of Pandas is very easy.Install it using this command:
**pip install pandas**
If this command fails, then use a python distribution that already has Pandas installed like, Anaconda, Spyder etc.Once Pandas is installed, import it in your applications by adding the import keyword:
**import pandas**

Pandas can read 3 data structure - Series, dataframe, panel
Series is 1 D array and dataframe is 2D. if series is row then dataframe is whole table.

```python
import pandas as pd
pd.__version__           #Its a pandas version
```

```python
# create pandas series
list1=['sunday','monday','tuesday','wed','thurs','fri','sat']
```

```python
s1=pd.Series(list1)
print(s1)
```

```python
s2=pd.Series(list1,index=[100,101,102,103,104,105,106])      # you can change index numbers
print(s2)
```

```python
s2=pd.Series(list1,index=[100,101,102,103,104,105,106],name='myvalues')      # you can give name to column
print(s2)
```

```python
s2[103]            # access values by giving index num
```

```python
s2[0:3]          #slicing of index upto 3 slices i.e.0,1,2 index
```

```python
s3=pd.Series(24,index=[100,101,102,103,104,105,106])     
print(s3)
```

```python
s4=pd.Series([24,34,78,12,11,45],index=[100,101,102,103,104,105]) 
max(s4)               #check for max value
```

```python
min(s4)                #check for min value
```

```python
s4[s4>35]           #print values greater than 35
```

```python
# pandas Join
df1 = pd.DataFrame({'A': [1,2,3],
                   'B': [10,20,30]})
 
 
df2 = pd.DataFrame({'C': [4,5,6],
                   'D': [40,50,60]})
 
display(df1, df2)
```

```python
df1.join (df2)
```

```python
df2.join (df1)
```

```python
#index values should be same otherwise NaN values shown at output
df1 = pd.DataFrame({'A': [1,2,3],
                   'B': [10,20,30]},
                   index = ['a','b','c'])
 
df2 = pd.DataFrame({'C': [4,5,6],
                   'D': [40,50,60]},
                  index = ['5','6','7'])
 
df2.join (df1)
```

```python
# now values are differant in both dataframes. 3 values and 2 values
df1 = pd.DataFrame({'A': [1,2,3],
                   'B': [10,20,30]},
                   index = ['a','b','c'])
 
df2 = pd.DataFrame({'C': [4,5],
                   'D': [40,50]},
                  index = ['a','b'])
 
display(df1, df2)
```

```python
#It joins according to right side df2 having 2 values. so it print 2 values.U can also use inner insted of right
df1.join(df2, how = 'right')
```

```python
#It joins according to left side df1 having 3 values. so it print 3 values. U can also use outer insted of left
df1.join(df2, how = 'left')
```
